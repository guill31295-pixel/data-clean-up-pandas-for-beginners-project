#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# TODO
mayor_prec = ds['price'].max()

ind_mayor_prec = ds['price'].idxmax()

row_info = ds.index[ind_mayor_prec]

      
print (f"La casa con direccion {ds.iloc[row_info,12]}, {ds.iloc[row_info,20]}, {ds.iloc[row_info,16]}, es la mas cara y su precio es {mayor_prec}")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# TODO
menor_prec = ds['price'].min()

ind_menor_prec = ds['price'].idxmin()

row_info = ds.index[ind_menor_prec]

print (f"La casa con direccion {ds.iloc[row_info,12]}, {ds.iloc[row_info,20]}, {ds.iloc[row_info,16]}, es la mas barata y su precio es {menor_prec}")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
prec.hist()
plt.xlabel("Precios")
plt.ylabel("Frecuencia")
plt.title("Histograma de precios de Arroyomolinos (Madrid)")
plt.show() 

# Trazar el histograma de precios
```

Se puede observar en la imagen anterior una distribucion normal de datos en la cual aproximadamente el 95% de precios se encuentran rondando los 200.000 y 300.000 euros, mientras que un menor porcentaje de precios (aproximadamente el 2%) ronda entre 0 y 180.000 euros para los percentiles inferiores y el porcentaje restante desde 420.000 hasta 600.000 euros para los percentiles superiores.
